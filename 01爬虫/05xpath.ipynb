### xpath解析
- pip install lxml
- xpath解析原理
    1. 实例化etree类型对象，且将页面源码数据加载到对象中
    2. 需要调用该对象的xpath方法结合不同形式的xpath表达式进行标签定位和数据提取
- etree对象实例化
    - etree.parse(fileName)
    - etree.HTML(page_text)
- xpath()方法返回的是一个列表

### 标签定位
- 在xpath表达式中，最左侧的/表示含义：当前定位标签必须从根节点开始进行定位
- xpath表达式中，最左侧的//表示含义：从任意位置进行标签定位
- xpath表达式中，非最左侧的//表示含义：多个层级的意思
- 属性定位：//tag[@attrName="value"]
- 索引定位：//tag[index]：index为索引，index从1开始

### 提取数据
- 取文本：
    - /text()：标签的子文本
    - //text()：标签的所有文本
    
- 取属性：
    - /tag/@attrName

```python
from lxml import etree

tree = etree.parse("./test.html")
tree.xpath("/html/head/meta")[0]   # 绝对路径
tree.xpath("//meta")[0]  # 相对路径，将整个页面源码中所有meta进行定位
tree.xpath("/html//meta")[0]
# 属性定位
tree.xpath("//div[@class='song']")
# 索引定位
tree.xpath("//div[3]")   # 索引从1开始

# 取文本
tree.xpath("//div[@class='song']/text()")    # /text()：标签的子文本
tree.xpath("//div[@class='song']//text()")   # //text()：标签的所有文本

# 取属性
tree.xpath("//a[1]/@href")
```

- 另一种较为通用的xpath表达式使用形式：管道符在xpath表达式中的应用，用于解析不规则布局页面数据
    - 爬取糗事百科内容：作者名称和段子内容（https://www.qiushibaike.com/text/page/2/）

```python
from lxml import etree

url = r"https://www.qiushibaike.com/text/page/2/"
headers = {
    "User-Agent":r"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_7_0) AppleWebKit/535.11 (KHTML, like Gecko) Chrome/17.0.963.56 Safari/535.11"
}
page_text = requests.get(url,headers=headers).text
tree = etree.HTML(page_text)
divs = tree.xpath("//div[@class='col1 old-style-col1']/div")
for div in divs:
    author = div.xpath("./div[1]/a[2]/h2/text() | .div[1]/span[2]/h2/text()")[0]
    content_href = r"https://www.qiushibaike.com" + div.xpath("./a/@href")[0]
    content_text = requests.get(content_href,headers).text
    content_tree = etree.HTML(content_text)
    content = content_tree.xpath("//div[@class='content']/text()")[0]
    print(author)
    print(content)
```

### 中文乱码问题
- 爬取网址：https://pic.netbian.com/4kmeinv/
- 要求：爬取图片和图片名称

```python
import os
'''
第1页：https://pic.netbian.com/4kmeinv/index.html
第2页：https://pic.netbian.com/4kmeinv/index_2.html
'''
# 爬取前2页
page = 0
for page in range(1,3):
    if(page==1):
        url = r"https://pic.netbian.com/4kmeinv/index.html"
    else:
        url = r"https://pic.netbian.com/4kmeinv/index_"+str(page)+".html"
    res = requests.get(url,headers=headers)
    #res.encoding = "utf8"
    page_text = res.text
    tree = etree.HTML(page_text)
    lis = tree.xpath(r"//*[@id='main']/div[3]/ul/li")
    for li in lis:
        imgName = li.xpath("./a/b/text()")[0].encode("iso-8859-1").decode("gbk")
        imgSrc = r"https://pic.netbian.com" + li.xpath("./a/img/@src")[0]
        img = requests.get(imgSrc,headers).content
        if(not os.path.exists(r"./imgMeinv")):
            os.mkdir("./imgMeinv")
        with open("./imgMeinv/"+imgName+".jpg","wb") as f:
            f.write(img)
print("爬取完成")
```
