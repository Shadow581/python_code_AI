### matplotlib
- 将数据进行可视化，更直观的呈现；数据更加客观、更具有说服力
- 概念：最流行的python底层绘图库，主要做数据可视化图表

```python
from matplotlib import pyplot as plt
```

- 写在前面：解决x轴，y轴乱码方式

```python
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 有时候，字体更改后，会导致坐标轴中的部分字符无法正常显示，此时需要更改axes.unicode_minus参数
mpl.rcParams["axes.unicode_minus"] = False
```

- plt.plot(x,y)：传入x和y，通过plot绘制出折线图
- plt.show()：展示图形

```python
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,27,22,18,15,11]
# 绘图
plt.plot(x,y)
# 展示图片
plt.show()
```

- 设置图片大小：plt.figure(figsize=(width,height),dpi=)
    - 参数figsize：元组，宽和高
    - 参数dpi：图片的分辨率

```python
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,27,22,18,15,11]
# 设置图片大小
plt.figure(figsize=(20,8),dpi=80)

plt.plot(x,y)
plt.show()
```

- 保存图片：plt.savefig()，可以使用.svg保存矢量图，这样就不会失帧

- plt.xticks(ticks,labels)
    - 设置x轴
    - 参数ticks：设置刻度；参数labels：需要先设置参数ticks后，再进行设置参数labels（即将坐标轴x上ticks刻度替换成labels，基本上替换为中文）
- plt.yticks(ticks,labels)：设置y轴

```python
x = range(2,26,2)
y = [15,13,14.5,17,20,25,26,27,22,18,15,11]
plt.plot(x,y)
# 设置x轴、y轴刻度
plt.xticks(range(2,26))
plt.yticks(range(10,30,2))
plt.show()
```

- 练习：绘制10点至12点的气温（每一分钟）

- 添加描述信息
    - plt.xlabel(label)：设置x轴标题为label
    - plt.ylabel(label)：设置y轴标题为label
    - plt.title(label)：设置图片标题为label

```python
import random
import matplotlib

x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
plt.plot(x,y)

plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("The temperature at some point")

plt.show()
```

- 散点图：scatter()

```python
y_3 = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22,23]
y_10 = [26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,15,11,15,5,13,17,10,11,13,12,13,6,21]
x_3 = range(1,32)
x_10 = range(51,82)

x = list(x_3)+list(x_10)

plt.figure(figsize=(20,8),dpi=80)

plt.scatter(x_3,y_3,label="Mar")
plt.scatter(x_10,y_10,label="Oct")

xlabels = []
for i in x:
    if i <= 31:
        xlabels.append("3月"+str(i)+"日")     # 月、日字样，出现乱码，最开始有讲解解决乱码方式
    else:
        xlabels.append("10月"+str(i-50)+"日")

# 添加图例
plt.legend()
        
plt.xticks(ticks=x[::3],labels=xlabels[::3])


plt.show()
```

- 条形图：plt.bar()

```python
a = ["战狼2","速度与激情8","功夫瑜伽","西游记伏妖篇","变形金刚5：最后的骑士"]
b = [56.01,26.94,17.53,16.49,15.45]
plt.figure(figsize=(20,8),dpi=80)
plt.bar(range(len(a)),b,width=0.15,color=['r','b','y','g','orange'])   # width：条形图宽度
plt.xticks(range(len(a)),labels=a,rotation=30)
plt.show()
```

- 绘制横着的条形图：plt.barh()

```python
a = ["战狼2","速度与激情8","功夫瑜伽","西游记伏妖篇","变形金刚5：最后的骑士"]
b = [56.01,26.94,17.53,16.49,15.45]
plt.figure(figsize=(20,8),dpi=80)
plt.barh(range(len(a)),b,height=0.15,color="orange")  # height：条形图宽度
plt.yticks(range(len(a)),labels=a)
plt.show()
```

```python
a = ["猩球崛起3：终极之战","敦刻尔克","蜘蛛侠：英雄归来","战狼2"]
b_14 = [2358,399,2358,262]
b_15 = [12357,156,2045,168]
b_16 = [15746,312,4497,319]

plt.figure(figsize=(20,8),dpi=80)

x_14 = list(range(len(a)))
x_15 = [i+0.2 for i in x_14]
x_16 = [i+0.2 for i in x_15]

plt.bar(range(len(a)),b_14,color="red",width=0.2,label="14号")
plt.bar(x_15,b_15,color="orange",width=0.2,label="15号")
plt.bar(x_16,b_16,color="pink",width=0.2,label="16号")

# 设置x轴刻度
plt.xticks(ticks=x_15,labels=a)

# 设置图例
plt.legend()

plt.show()
```
