#### 二、ndarray属性
- ndim：维度
- shape：形状
- size：总长度
- dtype：元素类型

```python
arr = np.random.randint(1,10,(3,4))
arr.shape    # (3, 4)
arr.size     # 12
arr.dtype    # dtype('int32')
arr.ndim     # 2
```

#### 三、ndarray基本操作
- 1.索引：一维数组与列表完全一致，多维数组同理

```python
arr[0][2]
```

- 2.切片

```python
# 获取二维数组前2行
arr[0:2]
# 获取二维数组前2列
# arr[行，列]
arr[:,0:2]
```

```python
# 获取二维数组前2行和前2列
arr[0:2,0:2]
```

```python
# 数组行倒序
arr[::-1]
```

```python
# 数组列倒序
arr[:,::-1]
```

```python
# 数组全部倒序
arr[::-1,::-1]
```

```python
# 将图片左右翻转
plt.imshow(imgArr[:,::-1])   # 下面写法也可
plt.imshow(imgArr[:,::-1,:])
```

- 4.级联：np.concatenate((arr1,arr2,...),axis=0,out)
    - (arr1,arr2,...)参数：将arr1,arr2等数组拼接在一起
    - axis参数：0表示列拼接，1表示行拼接
- 注意：
    - (arr1,arr2,...)中，维度必须相同，如：二维数组或三维数组等等
    - 形状相符：在维度一致情况下，若进行行级联，必须保证级联的数组行保持一致；若进行列级联，必须保证级联的数组列保持一致

- 1）一维、二维、多维数组的级联，实际操作中级联多为二维数组

```python
np.concatenate((arr,arr),axis=1)
```

#### 四、ndarray聚合操作

- 1.arr.sum(axis=None)：求和
    - 参数axis：0列求和，1行求和；若为None，则所有元素求和

```python
arr.sum(axis=1)
```

- 2.arr.max(axis=None),arr.min(axis=None)：求最大值、最小值
    - 参数axis：0列求最值，1行求最值；若为None，则所有元素求最值

```python
print(arr.max())
print(arr.max(axis=1))
print(arr.min(axis=0))
```

- 3.arr.mean(axis=None)：求平均值
    - 参数axis：0列求平均值，1行求平均值；若为None，则所有元素求平均值

```python
arr.mean()
```

#### 矩阵的运算
- np.matmul(A,B)
- np.dot(A,B)
- 区别：二者都是矩阵乘法。np.matmul中禁止矩阵与标量的乘法。在矢量乘矢量的內积运算中，np.matmul与np.dot没有区别。

```python
a = np.array([[1,2],[3,4],[7,9]])
b = np.array([[5,6],[8,2]])
np.dot(a,b)
```

```python
c = 100  # 标量
np.dot(a,c)
```

```python
np.matmul(a,c)    # 报错，np.matmul不支持矢量与标量进行运算
```
